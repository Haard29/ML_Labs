```python
# NLTK Exercise.
import nltk
from nltk.corpus import movie_reviews
import matplotlib.pyplot as plt
import random
nltk.download('movie_reviews')
movie_reviews.categories()
print("Positive movie reviews : ",len(movie_reviews.fileids('pos')))
print("Negative movie reviews : ",len(movie_reviews.fileids('neg')))
fig = plt.figure(figsize=(5, 5))
labels = 'Positives', 'Negatives'
sizes = [len(movie_reviews.fileids('pos')),len(movie_reviews.fileids('neg'))]
plt.pie(sizes, labels=labels, autopct='%1.1f%%',
shadow=False, startangle=90)
plt.axis('equal')
plt.show()
```

```python
nltk.download('stopwords')
import re
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

reviews = movie_reviews.raw('pos/cv944_13521.txt')
# print('\033[92m'+reviews)
review_tokens = reviews.split()
# print()
# print('\033[94m'+'Tokenized string:')
# print(review_tokens)

stopwords_english = stopwords.words('english')
# print('Stop words\n')
# print(stopwords_english)
# print('\nPunctuation\n')
# print(string.punctuation)

print('\033[92m')
print(review_tokens)
review_clean = []
for word in review_tokens:
  if( word not in stopwords_english and word not in string.punctuation ):
    review_clean.append(word)
print('\033[94m')
print('removed stop words and puntuations:\n')
print(review_clean)
print('\033[92m')
print(review_clean)


stemmer = PorterStemmer()
review_stem = []
for word in review_clean:
  stem_word = stemmer.stem(word)
  review_stem.append(stem_word)
print('\033[94m')
print('stemmed words:')
print(review_stem)
```

```python
nltk.download('wordnet')
print('\033[92m')
print(review_clean)

wordnet_lemmatizer = WordNetLemmatizer()

review_lem = []

for word in review_clean:
  lem_words = wordnet_lemmatizer.lemmatize(word)
  review_lem.append(lem_words)

print('\033[94m')
print('lemmatized words:')
print(review_lem)
```
